```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("C:\\Users\\Public\\Downloads\\fifa_data.csv")
```

```python
df
```

```python
#1.Which country has the most number of players
df['Nationality'].value_counts().idxmax()
# The country with most number of players is ENGLAND.
```

```python
#2.Plot a bar chart of 5 top countries with the most number of players.
MOST_PLAYER=df['Nationality'].value_counts().head(5)
plt.bar(MOST_PLAYER.index,MOST_PLAYER.values)
plt.title('Countries having most number of players')
plt.xlabel('Name of the Country')
plt.ylabel('Number of players')
plt.show

# Top 5 Countries with most number of players are shown above through bar chart.Out of which England is having the most number of players...
```

```python
#3.Which player has the highest salary?
MAX_SALARY=df.loc[df['Wage'].idxmax(), ['Name', 'Wage']]
MAX_SALARY
# F Marchetti is the player who is having highest salary which is about 9k Euro.
```

```python
#4.Plot a histogram to get the salary range of the players.
plt.hist(df['Wage'],color='b') 
plt.title('Salary Range of Players')
plt.xlabel('Name of the Player')
plt.ylabel('Salaries')
plt.show

# The above histogram shows the salary range of players.In this we can find there are players who are getting salary more than 9k Euro and some are getting very few only...
```

```python
#5.Who is the tallest player in the fifa?
Tall_Player=df.sort_values(by=['Height'],ascending=False)
Tall_Player[['Name','Height']].head(5)
#tallest players are T.Holy & D.Hodzic witha height of 6
```

```python
#5.Who is the tallest player in the fifa?
TALL_PLAYER= df.loc[df['HEIGHT'].idxmax(), 'Name']
TALL_PLAYER
```

```python
#6.Which club has the most number of players?
MOST_PLAYERS=df['Club'].value_counts()
MOST_PLAYERS
# There are total 5 countries with maximum number of players..and fc barca, valencia, fortuna,cardiff,rayo are having 33 number of players..
```

```python
#7.Which foot is most preferred by the players?Draw a bar chart for preferred foot
Pref_Foot=df['Preferred Foot'].value_counts()
plt.bar(Pref_Foot.index,Pref_Foot.values,color='b')
plt.title('Foot style of players')
plt.xlabel('Preffered Foot')
plt.ylabel('Number of players')
plt.show()
# Majority of players are right footed and left footed players are comparitively 1/3rd.
```
